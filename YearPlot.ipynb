```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns

df = pd.read_csv('./total.csv', names = ['year', 'week', 'SO2', 'NO2', 'CO', 'O3', 'PM2_5', 'PM10'])
df.head()
```

```python
for item in ('SO2', 'NO2', 'CO', 'O3', 'PM2_5', 'PM10'):
    g = sns.relplot(x='week', y=item, ci=None, kind = 'line', data=df)
    g.figure.autofmt_xdate()
```
