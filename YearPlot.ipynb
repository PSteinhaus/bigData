```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns

df = pd.read_csv('joined_data.csv', names = ['date', 'SO2 in ppm', 'NO2 in ppm', 'O3 in ppm', 'CO in ppm', 'PM2.5 in µg/m³', 'PM10 in µg/m³', 'mittlere Windgeschwindigkeit'])
df.head()
```

```python
fig, axes = plt.subplots(2,3, figsize=(23,10))
plt.rc('font', size=13) #controls default text size

i = 0
for item in ('SO2 in ppm', 'NO2 in ppm', 'CO in ppm', 'O3 in ppm', 'PM2.5 in µg/m³', 'PM10 in µg/m³'):
    ax_x = i % 3
    ax_y = i // 3
    sns.scatterplot(x = 'mittlere Windgeschwindigkeit', y = item, data = df, ax = axes[ax_y][ax_x], s = 1.5)
    i += 1

fig.suptitle('Belastung in Abhängigkeit der mittleren Windgeschwindigkeit in der Region', fontsize = 18)
fig.savefig("wind_scatter.png", dpi=150)
```

```python
fig, axes = plt.subplots(2,3, figsize=(23,10))
plt.rc('font', size=12) #controls default text size

i = 0
for item in ('SO2 in ppm', 'NO2 in ppm', 'CO in ppm', 'O3 in ppm', 'PM2.5 in µg/m³', 'PM10 in µg/m³'):
    ax_x = i % 3
    ax_y = i // 3
    g = sns.lineplot(x='Woche', y=item, ci=None, data=df, ax = axes[ax_y][ax_x], hue='Jahr', palette='Paired')
    i += 1

fig.suptitle('Mittelwerte pro Kalenderwoche', fontsize = 18)
fig.savefig("week_averages.png", dpi=150)
```
