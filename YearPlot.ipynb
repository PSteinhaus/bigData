```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns

df = pd.read_csv('./totalWeekAverages.csv', names = ['year', 'week', 'SO2', 'NO2', 'CO', 'O3', 'PM2.5 in ppm', 'PM10'])
df.head()
```

```python
for item in ('SO2', 'NO2', 'CO', 'O3', 'PM2.5 in ppm', 'PM10'):
    g = sns.relplot(x='week', y=item, ci=None, kind = 'line', data=df, hue='year', palette='Paired')
    g.figure.autofmt_xdate()
```
